```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

output_notebook()
```

```python
from IPython.display import display

# Membaca dataset dari file Excel
file_path = 'Employee-Master.xlsx'  # Ganti dengan path file Excel Anda
df = pd.read_excel(file_path)

# Menampilkan beberapa baris pertama dari DataFrame dalam bentuk tabel
display(df)
```

```python
#VISUALISASI 1
marital_by_age = pd.DataFrame(df)

# Buat ColumnDataSource dari DataFrame
source = ColumnDataSource(marital_by_age)

# Membuat figure untuk bar chart
bar_chart = figure(x_range=marital_by_age['MaritalStatus'].astype(str).unique(),  # Mengasumsikan kolom 'Country' dan 'Sales' ada dalam dataset
                   title='Status Pernikahan Berdasarkan Umur',
                   x_axis_label='Status Pernikahan',
                   y_axis_label='Umur',
                   height=300)

# Menambahkan bar ke chart
bar_chart.vbar(x='MaritalStatus', top='Age', source=source, color='red', width=0.5)
bar_chart.y_range.start = 0

# Menampilkan chart
show(bar_chart)
```

```python
#VISUALISASI 2
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

gender_by_age = pd.DataFrame(df)

# Menghitung rata-rata umur berdasarkan Departemen
gender_by_age = df.groupby('Gender')['Age'].mean().reset_index()

# Buat ColumnDataSource dari DataFrame yang baru
source = ColumnDataSource(gender_by_age)

# Membuat figure untuk area chart
area_chart = figure(x_range=gender_by_age['Gender'].astype(str).unique(),
                   title='Rata-rata Umur Berdasarkan Jenis Kelamin',
                   x_axis_label='Gender',
                   y_axis_label='Average Age',
                   height=300)

# Menambahkan area ke chart
area_chart.varea(x='Gender', y1=0, y2='Age', source=source, color='red')

# Menampilkan chart
show(area_chart)
```

```python
# VISUALISASI 3
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Assuming 'df' is your DataFrame
# Calculate average age by length of service
age_by_length = df.groupby('Length of Service')['Age'].mean().reset_index()

# Create a ColumnDataSource
source = ColumnDataSource(age_by_length)

# Create a figure for the line chart
line_chart = figure(title='Rata-rata Umur Berdasarkan Masa Kerja',
                    x_axis_label='Masa Kerja (Tahun)',
                    y_axis_label='Rata-rata Umur',
                    height=300)

# Add the line plot to the chart
line_chart.line(x='Length of Service', y='Age', source=source, line_width=2, color='red')

# Display the chart
show(line_chart)
```

```python
#VISUALISASI 4

import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

salary_by_length = pd.DataFrame(df)

# Menghitung rata-rata umur berdasarkan Departemen
salary_by_length = df.groupby('Length of Service')['Salary Monthly'].mean().reset_index()

# Buat ColumnDataSource dari DataFrame yang baru
source = ColumnDataSource(salary_by_length)

# Membuat figure untuk scatter plot
scatter_plot = figure(title='Rata-rata Gaji Bulanan Berdasarkan Lama Bekerja',
                   x_axis_label='Lama Bekerja',
                   y_axis_label='Gaji Bulanan',
                   height=300)

# Menambahkan scatter plot ke chart
scatter_plot.circle(x='Length of Service', y='Salary Monthly', source=source, size=8, color='red', alpha=0.8)

# Menampilkan chart
show(scatter_plot)
```

```python
#VISUALISASI 5
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

dept_by_salary = pd.DataFrame(df)

# Menghitung total gaji bulanan berdasarkan Departemen
dept_by_salary = df.groupby('Dept')['Salary Monthly'].sum().reset_index()

# Buat ColumnDataSource dari DataFrame yang baru
source = ColumnDataSource(dept_by_salary)

# Membuat figure untuk line chart
# Create a second line chart
line_chart2 = figure(x_range=dept_by_salary['Dept'].astype(str).unique(),
                     title='Total Gaji Bulanan Tiap Departemen',
                     x_axis_label='Dept',
                     y_axis_label='Total Salary Monthly',
                     height=300)

# Add the line to the second chart
line_chart2.line(x='Dept', y='Salary Monthly', source=source, line_width=2, line_color='red')

# Include both line charts in the dashboard
dashboard = column(bar_chart, area_chart, line_chart, scatter_plot, line_chart2)

# Menampilkan chart
show(line_chart)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Assuming 'df' is your DataFrame
# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a figure for the histogram
hist, edges = np.histogram(df['Age'], density=True, bins=20)
hist_plot = figure(title='Age Distribution of Employees',
                   x_axis_label='Age',
                   y_axis_label='Frequency',
                   height=300)

# Add the histogram to the chart
hist_plot.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
            fill_color="red", line_color="white", alpha=0.9)

# Display the chart
show(hist_plot)
```

```python
from bokeh.layouts import column, row
from bokeh.models import Div

# Create a title Div
title = Div(text="<h1>Employee Data Dashboard</h1>", width=800)
# Arrange plots in two columns
column1 = column(bar_chart, line_chart, area_chart)
column2 = column(scatter_plot, line_chart2, hist_plot)
# Arrange columns horizontally with the title
dashboard = column(title, row(column1, column2))

# Display the dashboard
show(dashboard)
```
